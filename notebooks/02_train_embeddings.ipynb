# 02 Train Embeddings

This notebook trains one Word2Vec Skip-gram model for each time period.

```python
from pathlib import Path
import sys

import pandas as pd

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.append(str(PROJECT_ROOT))

from src.preprocessing import sentences_by_period
from src.train_embeddings import save_models, train_models_by_period
```

```python
PROCESSED_PATH = PROJECT_ROOT / "data" / "processed" / "corpus_tokenized.pkl"
MODEL_DIR = PROJECT_ROOT / "models"

processed = pd.read_pickle(PROCESSED_PATH)
period_sentences = sentences_by_period(processed)
{period: len(sentences) for period, sentences in period_sentences.items()}
```

```python
models = train_models_by_period(
    period_sentences,
    vector_size=100,
    window=5,
    min_count=5,
    sg=1,
    workers=4,
    epochs=10,
    seed=42,
)

{period: len(model.wv) for period, model in models.items()}
```

```python
save_models(models, MODEL_DIR)
sorted(path.name for path in MODEL_DIR.glob("word2vec_*.model"))
```
